```python
import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Generate a random list of numbers to sort
random_list = [random.randint(0, 1000) for _ in range(5000)] 

# Record the start time
start_time = time.time() 
starting_time = time.time_ns()

# Execute the sorting algorithm
sorted_list = bubble_sort(random_list.copy())  # Use .copy() to avoid modifying the original list

# Record the end time
end_time = time.time() 
ending_time = time.time_ns()
# Calculate the elapsed time
elapsed_time = end_time - start_time 

print(f"Start time: {start_time}") 
print(f"Starting time: {starting_time}") 
print(f"End time: {end_time}")
print(f"Ending time: {ending_time}")
print(f"Execution time: {elapsed_time:.6f} seconds") 
```

To measure and print the start and end times of a sorting algorithm in milliseconds or nanoseconds, you can utilize the
time module, specifically time.time_ns() and time.perf_counter_ns() for nanosecond precision. 
Here's an example:
python

import time
import random

def bubble_sort(arr):
    n = len(arr)
    for i in range(n - 1):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

# Generate a random list of numbers to sort
random_list = [random.randint(0, 1000) for _ in range(5000)]

# --- Using time.time_ns() for nanoseconds (Python 3.7+) ---
start_time_ns = time.time_ns()  # Get current time in nanoseconds
sorted_list_time_ns = bubble_sort(random_list.copy())
end_time_ns = time.time_ns()  # Get current time in nanoseconds

elapsed_time_ns = end_time_ns - start_time_ns
elapsed_time_ms_from_ns = elapsed_time_ns / 1_000_000 # Convert nanoseconds to milliseconds

print("--- Using time.time_ns() ---")
print(f"Start time (nanoseconds): {start_time_ns}")
print(f"End time (nanoseconds): {end_time_ns}")
print(f"Execution time: {elapsed_time_ns} nanoseconds")
print(f"Execution time: {elapsed_time_ms_from_ns:.6f} milliseconds")
print("-" * 30)

# --- Using time.perf_counter_ns() for high-resolution nanoseconds ---
start_time_perf_ns = time.perf_counter_ns() # Get high-resolution time in nanoseconds
sorted_list_perf_ns = bubble_sort(random_list.copy())
end_time_perf_ns = time.perf_counter_ns() # Get high-resolution time in nanoseconds

elapsed_time_perf_ns = end_time_perf_ns - start_time_perf_ns
elapsed_time_ms_from_perf_ns = elapsed_time_perf_ns / 1_000_000 # Convert nanoseconds to milliseconds

print("--- Using time.perf_counter_ns() ---")
print(f"Start time (nanoseconds): {start_time_perf_ns}")
print(f"End time (nanoseconds): {end_time_perf_ns}")
print(f"Execution time: {elapsed_time_perf_ns} nanoseconds")
print(f"Execution time: {elapsed_time_ms_from_perf_ns:.6f} milliseconds")

Use code with caution.
Explanation

    time.time_ns(): This function, available in Python 3.7 and later, returns the current time in nanoseconds since the epoch as an integer value.
    time.perf_counter_ns(): This function, also available from Python 3.7, provides a high-resolution performance counter in nanoseconds, according to GeeksforGeeks. It's ideal for measuring short durations and is not affected by system clock changes, notes ZetCode.
    Conversion to Milliseconds: There are 1,000,000 nanoseconds in a millisecond; therefore, dividing the nanosecond duration by 1,000,000 provides the elapsed time in milliseconds.
    Printing: The start and end timestamps, along with the calculated elapsed time in both nanoseconds and milliseconds, are printed using f-strings for clear output. 

Note: For benchmarking code, time.perf_counter() or time.perf_counter_ns() are generally preferred over time.time() or time.time_ns() as they offer higher resolution and are not subject to system time adjustments, says Built In.
