```python
from PIL import Image
path = r"D:\Python Code\071_DIP\image.png"
# Open an image file
img = Image.open(path)
```

```python
# Rotate
img.rotate(45, expand=True)
```

```python
#Scale
# Define the new size (width, height)
new_size = (500, 300)
# Resize the image
img.resize(new_size)
```

```python
# Flip
img.transpose(Image.ROTATE_180)
```
