```python
arr = [1, 4, 3, 2, 0, 2, 3, 5]
len_ = len(arr)
step_count = {"comparision": 0, "swaps": 0}
```

### Bubble Sort

```python
for i in range(len_):
    
    for j in range(0, len_ - i - 1):
        step_count["comparision"] += 1
        if arr[j] > arr[j + 1]:
            arr[j + 1], arr[j] = arr[j], arr[j + 1]
            step_count["swaps"] += 1       
    print(arr)
    i, j  = i + 1, j + 1

print("Input size: ", len_)
print(f"Counts: {step_count}")
```

### Selection Sort

```python
'''Assume the current position holds the minimum element'''
for i in range(0, len_): 
    min = arr[i]
    j = i - 1
    for j in range(i + 1, len_):
        step_count["comparision"] += 1
        if arr[j] < arr[min]:
            min = j
            step_count["swaps"] += 1
    arr[i], arr[min] = arr[min], arr[i]
    print(arr)

print("Input size: ", len_)
print(f"Counts: {step_count}")
```

### Insertion Sort

```python
for i in range(1, len_):
    key = arr[i]
    j = i - 1
    step_count["comparision"] += 1
    while(j >= 0 and key < arr[j]):
        arr[j + 1] = arr[j] 
        j = j - 1 
        arr[j + 1] = key
        step_count["swaps"] += 1
        print(arr)  

print("Input size: ", len_)
print(f"Counts: {step_count}")    
```

To count the frequency of each step in a sorting algorithm in Python, you need to instrument the algorithm's code to increment a counter each time a specific operation or "step" occurs. The definition of a "step" will depend on what you want to measure (e.g., comparisons, swaps, assignments).
Here's a general approach using a dictionary to store the counts of different step types:

    Define a Counter Dictionary: Initialize an empty dictionary, for example, step_counts = {}, to store the frequencies of different operations.
    Instrument the Sorting Algorithm: Modify the sorting algorithm's code to increment the relevant counter whenever a specific operation occurs.
        Comparisons: Increment a "comparisons" counter every time two elements are compared.
        Swaps: Increment a "swaps" counter every time two elements are exchanged.
        Assignments: Increment an "assignments" counter every time a value is assigned to a variable (especially relevant for algorithms like Counting Sort).
        Function Calls/Recursion: Increment a counter for each recursive call or function invocation. 
    Example (Bubble Sort with step counting):

    def bubble_sort_with_steps(arr):

    Analyze Results: After running the instrumented algorithm, the step_counts dictionary will contain the frequency of each defined step, allowing you to analyze the algorithm's performance in terms of specific operations.

Considerations:

    Granularity of "Steps":
    Decide what level of detail you need for your "steps." Do you want to count every single assignment, or just major operations like comparisons and swaps?
    Algorithm Specifics:
    The types of steps you track will vary depending on the sorting algorithm. For example, Counting Sort involves counting occurrences and cumulative sums, while Merge Sort involves merging operations.
    Overhead:
    Adding counters will introduce a small amount of overhead to the algorithm's execution time, but for analysis purposes, this is generally acceptable.

## Merge Sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if(left_half[i] < right_half[j]):
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[i]
                j += 1
            k += 1
        
        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1
            
        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1

array = [38, 23, 56, 3, 42, 38]
merge_sort(array)
print(array)
```

```python
import time
def measure_time(arr):
    start_time = time.time()
    merge_sort(arr)
    end_time = time.time()
    return end_time - start_time

# Test Cases
n = 10000  # Size of the array

# Best Case: Already sorted array
best_case = list(range(n))
best_time = measure_time(best_case)

# Average Case: Random array
import random
average_case = [random.randint(0, n) for _ in range(n)]
average_time = measure_time(average_case)

# Worst Case: Reverse sorted array
worst_case = list(range(n, 0, -1))
worst_time = measure_time(worst_case)

# Output Results
print(f"Best Case Time: {best_time:.6f} seconds")
print(f"Average Case Time: {average_time:.6f} seconds")
print(f"Worst Case Time: {worst_time:.6f} seconds")
```
