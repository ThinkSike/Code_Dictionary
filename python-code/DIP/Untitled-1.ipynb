```python
# Implementation of relationship between neighbor pixels.
img = i.open(r"C:\Users\DESPU\017_DIP\mu077elk.jpg")#.convert('L')
pixels = img.load()
x , y = 50 ,50
neighbors_by_4_Connect = [
        pixels[x, y - 1],  # Top
        pixels[x, y + 1],  # Bottom
        pixels[x - 1, y],  # Left
        pixels[x + 1, y]  # Right
]

neighbors_by_8_Connect = [
        pixels[x - 1, y - 1], pixels[x, y - 1], pixels[x + 1, y - 1],  # Top row
        pixels[x - 1, y    ]                  , pixels[x + 1, y    ],  # Middle row
        pixels[x - 1, y + 1], pixels[x, y + 1], pixels[x + 1, y + 1]   # Bottom row
]

all_neighbors = [neighbors_by_4_Connect + neighbors_by_8_Connect]

print("Pixel: ", pixels[x, y])
print("Neighbors adjacent: :", neighbors_by_4_Connect)
print("Neighbors diagonal: ", neighbors_by_8_Connect)
print("All_neighbors: ", all_neighbors)
pixels

pixel1=neighbors_by_4_Connect
pixel2=neighbors_by_8_Connect
pixel3=all_neighbors

img1=i.new("RGB",(3, 3))
img2=i.new("RGB",(3, 3))
img3=i.new("RGB",(3, 3))

img1.putpixel(pixel1)
img2.putpixel(pixel2)
img3.putpixel(pixel3)

img1 = img1.resize((300, 300), resample = i.NEAREST)
img2 = img2.resize((300, 300), resample = i.NEAREST)
img3 = img3.resize((300, 300), resample = i.NEAREST)

img1.show()
img2.show()
img3.show()
```
