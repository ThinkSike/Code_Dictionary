```python
# Q1 Design a python program using class and object for bank management? 
# Problem Statement: Create a BankAccount class that allows users to deposit, withdraw, and check their balance.
# The class should keep track of the account holder's name and the account balance. 
# Implement methods for depositing money, withdrawing money (with checks for sufficient balance), and displaying the current balance.
class BankAccount:
    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew: ${amount:.2f}. New balance: ${self.balance:.2f}")
            else:
                print("Insufficient balance for this withdrawal.")
        else:
            print("Withdrawal amount must be positive.")

    def check_balance(self):
        print(f"Current balance: ${self.balance:.2f}")
```

```python
# Q2 Design a python program using class and object for Library Management System?
# Problem Statement: Design a Book class that represents a book with attributes like title, author, and ISBN.
# Create a Library class that maintains a list of books available in the library. 
# Implement methods to add a book, remove a book, search for a book by title or author, and display all books in the library. 
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

    def __str__(self):
        return f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}"
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Added book: {book}")

    def remove_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print(f"Removed book: {book}")
                return
        print("Book not found.")

    def search_by_title(self, title):
        found_books = [book for book in self.books if book.title.lower() == title.lower()]
        if found_books:
            for book in found_books:
                print(book)
        else:
            print("No books found with that title.")

    def search_by_author(self, author):
        found_books = [book for book in self.books if book.author.lower() == author.lower()]
        if found_books:
            for book in found_books:
                print(book)
        else:
            print("No books found by that author.")

    def display_books(self):
        if not self.books:
            print("No books available in the library.")
        else:
            for book in self.books:
                print(book)
```

```python
# Q4 Design a python program using class and object for E-commerce Order System? 
# Problem Statement: Develop a Product class with attributes like product ID, name, price, and stock quantity.
# Create an Order class that allows users to add products to their order, calculate the total cost, and check out by reducing the stock quantity.
class Product:
    def __init__(self, product_id, name, price, stock_quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock_quantity = stock_quantity

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Price: ${self.price:.2f}, Stock: {self.stock_quantity}"
class Order:
    def __init__(self):
        self.products = []

    def add_product(self, product, quantity):
        if quantity > 0 and quantity <= product.stock_quantity:
            self.products.append((product, quantity))
            print(f"Added {quantity} of {product.name} to the order.")
        else:
            print("Invalid quantity or insufficient stock.")

    def calculate_total(self):
        total = sum(product.price * quantity for product, quantity in self.products)
        print(f"Total cost of the order: ${total:.2f}")
        return total

    def checkout(self):
        for product, quantity in self.products:
            product.stock_quantity -= quantity
            print(f"Checked out {quantity} of {product.name}. Remaining stock: {product.stock_quantity}")
        self.products.clear()
        print("Order completed.")
```

```python
# Q5 Design a python program using class and object for Car Rental System? 
# Problem Statement: Create a Car class with attributes such as make, model, year, and rental price per day. 
# Develop a RentalService class that manages the rental of cars. 
# Implement methods to rent a car, return a car, and calculate the total rental cost based on the number of days rented.
class Car:
    def __init__(self, make, model, year, rental_price_per_day):
        self.make = make
        self.model = model
        self.year = year
        self.rental_price_per_day = rental_price_per_day

    def __str__(self):
        return f"{self.year} {self.make} {self.model} - ${self.rental_price_per_day:.2f}/day"
class RentalService:
    def __init__(self):
        self.cars = []
        self.rented_cars = {}

    def add_car(self, car):
        self.cars.append(car)
        print(f"Added car: {car}")

    def rent_car(self, car, days):
        if car in self.cars and days > 0:
            self.cars.remove(car)
            self.rented_cars[car] = days
            print(f"Rented {car} for {days} days.")
        else:
            print("Car not available or invalid rental period.")

    def return_car(self, car):
        if car in self.rented_cars:
            days = self.rented_cars.pop(car)
            self.cars.append(car)
            print(f"Returned {car}. Total rental days: {days}")
        else:
            print("This car was not rented from us.")

    def calculate_total_cost(self, car):
        if car in self.rented_cars:
            days = self.rented_cars[car]
            total_cost = car.rental_price_per_day * days
            print(f"Total rental cost for {car}: ${total_cost:.2f}")
            return total_cost
        else:
            print("This car is not currently rented.")
            return 0.0
```

```python
# Q6 Design a python program using class and object for Inventory Management System? 
# Problem Statement: Design a Product class with attributes like name, price, and quantity in stock. 
# Create an Inventory class that allows adding new products, updating stock, and removing products.
# Implement methods to check stock levels, calculate the value of the inventory, and generate a restock report.
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def __str__(self):
        return f"Product: {self.name}, Price: ${self.price:.2f}, Quantity: {self.quantity}"
class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"Added product: {product}")

    def update_stock(self, name, quantity):
        for product in self.products:
            if product.name == name:
                product.quantity += quantity
                print(f"Updated stock for {product.name}. New quantity: {product.quantity}")
                return
        print("Product not found.")

    def remove_product(self, name):
        for product in self.products:
            if product.name == name:
                self.products.remove(product)
                print(f"Removed product: {product.name}")
                return
        print("Product not found.")

    def check_stock_levels(self):
        if not self.products:
            print("No products in inventory.")
        else:
            for product in self.products:
                print(product)

    def calculate_inventory_value(self):
        total_value = sum(product.price * product.quantity for product in self.products)
        print(f"Total inventory value: ${total_value:.2f}")
        return total_value

    def generate_restock_report(self, threshold):
        restock_items = [product for product in self.products if product.quantity < threshold]
        if restock_items:
            print("Products that need restocking:")
            for item in restock_items:
                print(item)
        else:
            print("All products are sufficiently stocked.")
```

```python
# Q7 Design a python program using class and object for online learning platform? 
# Problem Statement: Build a Course class with attributes such as course name, instructor, and a list of enrolled students. 
# Create a Student class with attributes like name, email, and a list of courses they are enrolled in. 
# Implement methods to enroll students in courses, drop courses, and display the courses a student is enrolled in.
class Course:
    def __init__(self, course_name, instructor):
        self.course_name = course_name
        self.instructor = instructor
        self.enrolled_students = []

    def enroll_student(self, student):
        if student not in self.enrolled_students:
            self.enrolled_students.append(student)
            student.courses.append(self)
            print(f"Enrolled {student.name} in {self.course_name}.")
        else:
            print(f"{student.name} is already enrolled in {self.course_name}.")

    def drop_student(self, student):
        if student in self.enrolled_students:
            self.enrolled_students.remove(student)
            student.courses.remove(self)
            print(f"Dropped {student.name} from {self.course_name}.")
        else:
            print(f"{student.name} is not enrolled in {self.course_name}.")

    def display_enrolled_students(self):
        if not self.enrolled_students:
            print("No students enrolled in this course.")
        else:
            print(f"Students enrolled in {self.course_name}:")
            for student in self.enrolled_students:
                print(student.name)
```
