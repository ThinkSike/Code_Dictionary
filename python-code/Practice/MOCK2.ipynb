```python
# If-Else: Accept three angles and determine whether they can form a valid triangle and identify its type.
def triangle_type(angle1, angle2, angle3):
    if angle1 <= 0 or angle2 <= 0 or angle3 <= 0:
        return "Invalid angles"
    if angle1 + angle2 + angle3 != 180:
        return "Not a triangle"
    if angle1 < 90 and angle2 < 90 and angle3 < 90:
        return "Acute triangle"
    elif angle1 == 90 or angle2 == 90 or angle3 == 90:
        return "Right triangle"
    else:
        return "Obtuse triangle"
```

```python
# Loop: Generate and display the Hailstone sequence starting from a given number.
def hailstone_sequence(n):
    sequence = []
    while n != 1:
        sequence.append(n)
        if n % 2 == 0:
            n //= 2
        else:
            n = 3 * n + 1
    sequence.append(1)
    return sequence
```

```python
# Function: Define a function to calculate compound interest over a given period.
def compound_interest(principal, rate, time):
    return principal * (1 + rate / 100) ** time
```

```python
# If-Else: Check if the given 5-digit number is an Armstrong number and a palindrome simultaneously.
def is_armstrong_and_palindrome(num):
    if num < 10000 or num > 99999:
        return False
    str_num = str(num)
    sum_of_powers = sum(int(digit) ** 5 for digit in str_num)
    is_armstrong = (sum_of_powers == num)
    is_palindrome = (str_num == str_num[::-1])
    return is_armstrong and is_palindrome
```

```python
# Loop: Generate a triangle pattern of prime numbers with n rows.
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
```

```python
# Function: Write a function to convert a Roman numeral to its integer equivalent.
def roman_to_integer(roman):
    roman_numerals = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
    }
    total = 0
    prev_value = 0
    for char in reversed(roman):
        value = roman_numerals[char]
        if value < prev_value:
            total -= value
        else:
            total += value
        prev_value = value
    return total
```

```python
# If-Else: Check if the given number is a spy number.
def is_spy_number(num):
    if num < 0:
        return False
    digit_sum = sum(int(digit) for digit in str(num))
    digit_product = 1
    for digit in str(num):
        digit_product *= int(digit)
    return digit_sum == digit_product
```

```python
# Loop: Print a right-angled triangle using numbers up to n.
def right_angled_triangle(n):
    triangle = []
    for i in range(1, n + 1):
        triangle.append(' '.join(str(j) for j in range(1, i + 1)))
    return '\n'.join(triangle)  
```

```python
# Function: Write a function to check whether a number is a Harshad number.
def is_harshad_number(num):
    if num <= 0:
        return False
    digit_sum = sum(int(digit) for digit in str(num))
    return num % digit_sum == 0
```

```python
# If-Else: Accept coordinates and determine which quadrant the point lies in.
def quadrant(x, y):
    if x > 0 and y > 0:
        return "Quadrant I"
    elif x < 0 and y > 0:
        return "Quadrant II"
    elif x < 0 and y < 0:
        return "Quadrant III"
    elif x > 0 and y < 0:
        return "Quadrant IV"
    else:
        return "On an axis or at the origin"
```

```python
# Loop: Display the first 10 Catalan numbers.
def catalan_numbers(n):
    catalan = [0] * (n + 1)
    catalan[0] = 1
    for i in range(1, n + 1):
        catalan[i] = sum(catalan[j] * catalan[i - 1 - j] for j in range(i))
    return catalan[:n]
catalan_numbers(10)  # Example usage
```

```python
# Loop: Print a hollow square pattern using asterisks.
def hollow_square(n):
    pattern = []
    for i in range(n):
        if i == 0 or i == n - 1:
            pattern.append('*' * n)
        else:
            pattern.append('*' + ' ' * (n - 2) + '*')
    return '\n'.join(pattern)
```

```python
# Function: Define a function that counts the frequency of each digit in a number.
def digit_frequency(num):
    frequency = {}
    for digit in str(num):
        if digit in frequency:
            frequency[digit] += 1
        else:
            frequency[digit] = 1
    return frequency
digit_frequency(1122334455)  # Example usage
```

```python
# If-Else: Accept a month and print the number of days in it (considering leap years).
def days_in_month(month, year):
    if month < 1 or month > 12:
        return "Invalid month"
    if month in [1, 3, 5, 7, 8, 10, 12]:
        return 31
    elif month in [4, 6, 9, 11]:
        return 30
    elif month == 2:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return 29
        else:
            return 28
    return "Invalid month"
```

```python
# Loop: Generate a pyramid of numbers.
def number_pyramid(n):
    pyramid = []
    for i in range(1, n + 1):
        row = ' '.join(str(j) for j in range(1, i + 1))
        pyramid.append(row.center(n * 2))
    return '\n'.join(pyramid)
number_pyramid(5)  # Example usage
```

```python
# Function: Write a function to calculate the sum of digits of a number until a single digit remains.
def sum_to_single_digit(num):
    while num >= 10:
        num = sum(int(digit) for digit in str(num))
    return num
```

```python
# If-Else: Accept a number and determine whether it&#39;s a Duck number.
def is_duck_number(num):
    if num < 0:
        return False
    str_num = str(num)
    return '0' in str_num and str_num[0] != '0'
```

```python
# Loop: Print all perfect numbers between 1 and 10000.
def perfect_numbers(limit):
    perfects = []
    for num in range(1, limit + 1):
        divisors_sum = sum(i for i in range(1, num) if num % i == 0)
        if divisors_sum == num:
            perfects.append(num)
    return perfects
perfect_numbers(1000)
```

```python
# Function: Create a function to check if a given string is a pangram.
def is_pangram(s):
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    return alphabet <= set(s.lower())
```

```python
# If-Else: Determine if a number is a Happy Number.
def is_happy_number(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit) ** 2 for digit in str(n))
    return n == 1
```

```python
# Loop: Generate Pascal’s triangle up to n rows.
def pascal_triangle(n):
    triangle = []
    for i in range(n):
        row = [1] * (i + 1)
        for j in range(1, i):
            row[j] = triangle[i - 1][j - 1] + triangle[i - 1][j]
        triangle.append(row)
    return triangle
pascal_triangle(5)  # Example usage, should return the first 5 rows of Pascal's triangle
```

```python
# Function: Write a function to find the second largest number in a list.
def second_largest(numbers):
    if len(numbers) < 2:
        return None
    first, second = float('-inf'), float('-inf')
    for number in numbers:
        if number > first:
            second = first
            first = number
        elif first > number > second:
            second = number
    return second if second != float('-inf') else None
```

```python
# If-Else: Accept a number and identify whether it is a Disarium number.
def is_disarium_number(num):
    str_num = str(num)
    total = sum(int(digit) ** (index + 1) for index, digit in enumerate(str_num))
    return total == num
```

```python
# Loop: Create an inverted pyramid pattern of stars.
def inverted_pyramid(n):
    pattern = []
    for i in range(n, 0, -1):
        pattern.append('*' * i)
    return '\n'.join(pattern)
```

```python
# Function: Write a function to calculate factorial recursively.
def factorial(n):
    if n < 0:
        return None  # Factorial is not defined for negative numbers
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)
```

```python
# If-Else: Check whether a given number is an Automorphic number.
def is_automorphic(num):
    square = num ** 2
    return str(square).endswith(str(num))
is_automorphic(25)  # Example usage, should return True since 25^2 = 625 ends with 25
```

```python
# Loop: Generate and print Floyd’s Triangle.
def floyds_triangle(n):
    triangle = []
    current_number = 1
    for i in range(1, n + 1):
        row = ' '.join(str(current_number + j) for j in range(i))
        triangle.append(row)
        current_number += i
    return '\n'.join(triangle)
```

```python
# Function: Define a function to check for a strong number.
def is_strong_number(num):
    if num < 0:
        return False
    digit_sum = sum(factorial(int(digit)) for digit in str(num))
    return digit_sum == num
```

```python
# If-Else: Validate a password based on length and presence of character types.
def is_valid_password(password):
    if len(password) < 8:
        return False
    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_special = any(c in '!@#$%^&*()-_=+[]{}|;:,.<>?/' for c in password)
    return has_upper and has_lower and has_digit and has_special
```

```python
# Loop: Display a butterfly pattern using stars.
def butterfly_pattern(n):
    pattern = []
    # Upper part
    for i in range(1, n + 1):
        pattern.append('*' * i + ' ' * (2 * (n - i)) + '*' * i)
    # Lower part
    for i in range(n, 0, -1):
        pattern.append('*' * i + ' ' * (2 * (n - i)) + '*' * i)
    return '\n'.join(pattern)
butterfly_pattern(5)
```

```python
# Function: Write a function to find the Nth prime number.
def nth_prime(n):
    if n < 1:
        return None
    count = 0
    num = 1
    while count < n:
        num += 1
        if is_prime(num):
            count += 1
    return num
```
