# Q1. WAP TO PERFORM ROTATION, SCALING, FLIPPING AND RESIZING ON AN IMAGE

```python
import cv2 
image = r"D:\071_python\image.png"
```

```python
# Read the image
img = cv2.imread(image)

# Display the image
cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Rotation
rotate_image = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2.imshow("Image", rotate_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Scaling
scale_x = 3
scale_y = 4
scale_image = cv2.resize(img, None, fx=scale_x, fy= scale_y, interpolation=cv2.INTER_LINEAR)
cv2.imshow("Image", scale_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Flipping
flip_image = cv2.flip(img, -2)
cv2.imshow("Image", flip_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Resizing
# (to specific dimension)
resize_image = cv2.resize(img, (300, 300))
cv2.imshow("Image", resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
