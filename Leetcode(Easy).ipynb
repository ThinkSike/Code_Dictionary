```python
def plusOne(digits):
    n = len(digits)
    for i in range(n-1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        digits[i] = 0
    return [1] + digits

digits = [1,2,9]
print(plusOne(digits))
```

Single One

```python
def singleNumber(nums):
    result = 0
    for num in nums:
        result ^= num
    return result

nums = [2,2,1]
print(singleNumber(nums))
```
