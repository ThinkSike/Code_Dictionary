Max Consecutive Ones

```python
def findMaxConsecutiveOnes(nums):
    i = 0
    consecutive = 0 
    max_consecutive = 0 
    for i in range(len(nums)):
        if nums[i] == 1:
            consecutive += 1
            max_consecutive = max(max_consecutive, consecutive)
        else:
            consecutive = 0
            i = i + 1
    return max_consecutive

nums = [1,1,0,1,1,1]
print(findMaxConsecutiveOnes(nums))
```

Even number of digits in a number

```python
nums = [12,345,2,6,7896]

count = sum(1 for num in nums if len(str(num)) % 2 == 0)
print(count)

def findNumbers(nums):
    count = 0
    for num in nums:
        digitCount = 0
        digits = num
        if digits == 0:
            digitCount = 1
        while digits > 0:
            digits //= 10
            digitCount += 1
        if digitCount % 2 == 0:
            count += 1
    return count 

print(findNumbers(nums))
```

Sorted squares

```python
def sortedSquares(nums):
    result = []
    for num in nums:
      number = num
      number **= 2
      result.append(number)
    result.sort()
    return(result)

nums = [-4,-1,0,3,10]
print(sortedSquares(nums))

result = [num ** 2 for num in nums]
result.sort()
print(result)
```

Duplicate zeros

```python
def duplicateZeros(arr):
    n = len(arr)
    zeros = arr.count(0)
    i = n - 1 # last index
    j = n + zeros - 1 # imaginary index
    while i >= 0: # start accessing no.s 1 by 1
        if j < n: # if no zeros
            arr[j] = arr[i] # end of extension = i
    
        if arr[i] == 0: # if found a zero
            j -= 1 # reduce 1 from extension demand
            if j < n: # if extension used, then place zero
                arr[j] = 0
        i -= 1
        j -=1
    return arr

arr = [1,0,2,3,0,4,5,0]
print(duplicateZeros(arr))
```

Merge Sorted Array

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

i = m - 1
j = n - 1
k = m + n - 1 # last index of merged array i.e 9
while i >= 0 and j >= 0: # inital num of both array
    if nums1[i] > nums2[j]: # if i[0] > j[0] then place i[0] at end of merged array
        nums1[k] = nums1[i] # 3 -> 9 
        i -= 1 
    else:
        nums1[k] = nums2[j] # 6 -> 9
        j -= 1 # reduce j to access next element of nums2
    k -= 1 # reduce k to place next element of merged array
while j >= 0: # if nums2 has remaining elements, place them in merged array
    nums1[k] = nums2[j] # place remaining element of nums2 in merged array
    j -= 1
    k -= 1
print(nums1)
```

Remove Element

```python
def removeElement(nums, val):
    i = 0
    for j in range(len(nums)): # access each element of nums
        if nums[j] != val: # if element is not val, place it at index i and increment i
            nums[i] = nums[j]  # place non-val element at index i
            i += 1
    return i

nums = [3,2,2,3]
val = 3
print(removeElement(nums, val))
```

Remove Duplicates

```python
def removeDuplicates(nums):
    if not nums:
        return 0
    i = 1  # slow pointer (next unique position)
# Compare current element with previous
    for j in range(1, len(nums)):
        if nums[j] != nums[j - 1]:
# If different → it's a new unique number
# Place it at index i
            nums[i] = nums[j]
            i += 1
    return i

def removeDuplicates_(nums):
    i = 0
    for x in nums:
        if i == 0 or x != nums[i - 1]:
            nums[i] = x
            i += 1
    return i

nums = [1,1,2]
print(removeDuplicates(nums))
print(removeDuplicates_(nums))
```

Check if exists

```python
def checkIfExist(arr):
    seen = set()
    
    for num in arr:
        # Check if double or half already exists
        if 2 * num in seen or (num % 2 == 0 and num // 2 in seen):
            return True
        seen.add(num)
    return False

arr = [10,2,5,3]
print(checkIfExist(arr))
```

Valid mountain array

```python
def validMountainArray(arr):
    n = len(arr)
    if n < 3:
        return False
    
    i = 0
    
    # Walk up
    while i + 1 < n and arr[i] < arr[i + 1]:
        i += 1
    
    # Peak can't be first or last
    if i == 0 or i == n - 1:
        return False
    
    # Walk down
    while i + 1 < n and arr[i] > arr[i + 1]:
        i += 1
    
    return i == n - 1

arr = [0,3,2,1]
print(validMountainArray(arr))
```

Replace elements with max_right

```python
def replaceElements(arr):
    n = len(arr)
    max_right = -1
    for i in range(n-1, -1, -1):
        current = arr[i]
        arr[i] = max_right
        max_right = max(max_right, current)
    return arr

arr = [17,18,5,4,6,1]
replaceElements(arr)
```

Move zeros

```python
nums = [0,1,0,3,12]

def MoveZeros(nums):
    i = 0
    for j in range(len(nums)):
        if nums[j] != 0:
            nums[i] = nums[j]
            i += 1
    for k in range(i, len(nums)):
        nums[k] = 0
    return nums

def moveZeroes(nums):
    j = 0  # position to place next non-zero

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[j], nums[i] = nums[i], nums[j]
            j += 1 
    return nums 

print(MoveZeros(nums))
print(moveZeroes(nums))
```

Sort Array by Parity

```python
def SortByParity(nums):
    i = 0
    for j in range(len(nums)):
        if nums[j] % 2 == 0:
            nums[i], nums[j] = nums[j], nums[i]
            i += 1
    return nums

nums = [3,1,2,4]
print(SortByParity(nums))
```

Height Checker

```python
def heightChecker(heights):
    expected = sorted(heights)
    return sum(h1 != h2 for h1, h2 in zip(heights, expected))

heights = [1,1,4,2,1,3]
print(heightChecker(heights))
```

Find Third max number in array

```python
def thirdMax(nums):
    first = second = third = -1
    for num in nums:
        if num > first:
            third = second
            second = first
            first = num
        elif first > num > second:
            third = second
            second = num
        elif second > num > third:
            third = num
    return third if third != -1 else first
    
nums = [3,2,1]
print(thirdMax(nums))
```

Find All Numbers Disappeared in an Array

```python
def findDissaperedNumbers(nums):
    n = len(nums)
    for num in nums:
        index = abs(num) - 1
        if nums[index] > 0:
            nums[index] = -nums[index]
    return [i + 1 for i in range(n) if nums[i] > 0]

nums = [4,3,2,7,8,2,3,1]
print(findDissaperedNumbers(nums))
```
