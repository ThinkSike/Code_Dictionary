Max Consecutive Ones

```python
class Solution:
    def findMaxConsecutiveOnes(self, nums: List[int]) -> int:
        i = 0
        consecutive = 0 
        max_consecutive = 0 
        for i in range(len(nums)):
            if nums[i] == 1:
                consecutive += 1
                max_consecutive = max(max_consecutive, consecutive)
            else:
                consecutive = 0
                i = i + 1
        return max_consecutive
```

```python
count = sum(1 for num in nums if len(str(num)) % 2 == 0)
print(count)

class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count = 0
        for num in nums:
            digitCount = 0
            digits = num
            if digits == 0:
                digitCount = 1
            while digits > 0:
                digits //= 10
                digitCount += 1
            if digitCount % 2 == 0:
                count += 1
        return count 
```

Sorted squares

```python
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        result = []
        for num in nums:
          number = num
          number **= 2
          result.append(number)
        result.sort()
        return(result)


# result = [num ** 2 for num in nums]
# result.sort()
# return(result)
```

Duplicate zeros

```python
class Solution:
    def duplicateZeros(self, arr: List[int]) -> None:
        n = len(arr)
        zeros = arr.count(0)
        i = n - 1 # last index
        j = n + zeros - 1 # imaginary index
        while i >= 0: # start accessing no.s 1 by 1
            if j < n: # if no zeros
                arr[j] = arr[i] # end of extension = i
        
            if arr[i] == 0: # if found a zero
                j -= 1 # reduce 1 from extension demand
                if j < n: # if extension used, then place zero
                    arr[j] = 0
            i -= 1
            j -=1
        return arr
```

Merge Sorted Array

```python
nums1 = [1,2,3,0,0,0]
m = 3
nums2 = [2,5,6]
n = 3

i = m - 1
j = n - 1
k = m + n - 1 # last index of merged array i.e 9
while i >= 0 and j >= 0: # inital num of both array
    if nums1[i] > nums2[j]: # if i[0] > j[0] then place i[0] at end of merged array
        nums1[k] = nums1[i] # 3 -> 9 
        i -= 1 
    else:
        nums1[k] = nums2[j] # 6 -> 9
        j -= 1 # reduce j to access next element of nums2
    k -= 1 # reduce k to place next element of merged array
while j >= 0: # if nums2 has remaining elements, place them in merged array
    nums1[k] = nums2[j] # place remaining element of nums2 in merged array
    j -= 1
    k -= 1
print(nums1)
```

Remove Element

```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        i = 0
        for j in range(len(nums)): # access each element of nums
            if nums[j] != val: # if element is not val, place it at index i and increment i
                nums[i] = nums[j]  # place non-val element at index i
                i += 1
        return i
```

Remove Duplicates

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if not nums:
            return 0
        i = 1  # slow pointer (next unique position)
    # Compare current element with previous
        for j in range(1, len(nums)):
            if nums[j] != nums[j - 1]:
    # If different → it's a new unique number
    # Place it at index i
                nums[i] = nums[j]
                i += 1
        return i

class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 0
        for x in nums:
            if i == 0 or x != nums[i - 1]:
                nums[i] = x
                i += 1
        return i
```

Check if exists

```python
class Solution:
    def checkIfExist(self, arr: List[int]) -> bool:
        seen = set()
        
        for num in arr:
            # Check if double or half already exists
            if 2 * num in seen or (num % 2 == 0 and num // 2 in seen):
                return True
            seen.add(num)
        
        return False
```

Valid mountain array

```python
class Solution:
    def validMountainArray(self, arr: List[int]) -> bool:
        n = len(arr)
        if n < 3:
            return False
        
        i = 0
        
        # Walk up
        while i + 1 < n and arr[i] < arr[i + 1]:
            i += 1
        
        # Peak can't be first or last
        if i == 0 or i == n - 1:
            return False
        
        # Walk down
        while i + 1 < n and arr[i] > arr[i + 1]:
            i += 1
        
        return i == n - 1
```
