```python
# Function: Write a function to find the GCD (Greatest Common Divisor) of two numbers using Euclidean algorithm.
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Example usage of the GCD function
num1 = int(input("Enter first number: "))   
num2 = int(input("Enter second number: "))
print(f"The GCD of {num1} and {num2} is {gcd(num1, num2)}")
```

```python
# If-Else: Take three side lengths and determine whether they can form a triangle. If yes, identify the type of triangle.
def triangle_type(a, b, c):
    if a + b > c and a + c > b and b + c > a:
        if a == b == c:
            return "Equilateral triangle"
        elif a == b or b == c or a == c:
            return "Isosceles triangle"
        else:
            return "Scalene triangle"
    else:
        return "Not a triangle"
```

```python
# Loop: Display the first 10 numbers of the Pell series.
def pell_series(n):
    pell_numbers = []
    a, b = 0, 1
    for _ in range(n):
        pell_numbers.append(a)
        a, b = b, 2 * b + a
    return pell_numbers
```

```python
# If-Else: Accept a number and print if it is a perfect square, perfect cube, both or neither.
import math
def check_perfect_square_and_cube(n):
    is_square = math.isqrt(n) ** 2 == n
    is_cube = round(n ** (1/3)) ** 3 == n
    
    if is_square and is_cube:
        return "Both perfect square and perfect cube"
    elif is_square:
        return "Perfect square"
    elif is_cube:
        return "Perfect cube"
    else:
        return "Neither perfect square nor perfect cube"
```

```python
# Loop: Find the factorial of a number without using any inbuilt function.
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

```python
# Function: Create a function to count the number of vowels, consonants, digits, and special characters in a string.
def count_characters(s):
    vowels = "aeiouAEIOU"
    counts = {
        'vowels': 0,
        'consonants': 0,
        'digits': 0,
        'special_characters': 0
    }
    
    for char in s:
        if char.isalpha():
            if char in vowels:
                counts['vowels'] += 1
            else:
                counts['consonants'] += 1
        elif char.isdigit():
            counts['digits'] += 1
        else:
            counts['special_characters'] += 1
            
    return counts
```

```python
# If-Else: Accept marks in five subjects and determine the grade using weighted average.
def calculate_grade(marks):
    if len(marks) != 5:
        return "Please provide marks for exactly five subjects."
    
    weighted_average = sum(marks) / 5
    
    if weighted_average >= 90:
        return "Grade A"
    elif weighted_average >= 75:
        return "Grade B"
    elif weighted_average >= 60:
        return "Grade C"
    elif weighted_average >= 50:
        return "Grade D"
    else:
        return "Grade F"    
```

```python
# Loop: Reverse a given integer number using loop.
def reverse_integer(n):
    reversed_num = 0
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
    return reversed_num
```

```python
# Function: Create a function to return the binary equivalent of a decimal number.
def decimal_to_binary(n):
    if n < 0:
        return "Negative numbers are not supported."
    binary_num = ""
    if n == 0:
        return "0"
    while n > 0:
        binary_num = str(n % 2) + binary_num
        n //= 2
    return binary_num
```

```python
# Loop: Display all twin prime pairs between 1 and 100.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def twin_primes(limit):
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    
    twin_prime_pairs = []
    for i in range(len(primes) - 1):
        if primes[i + 1] - primes[i] == 2:
            twin_prime_pairs.append((primes[i], primes[i + 1]))
    
    return twin_prime_pairs
# Example usage of twin primes function
twin_prime_pairs = twin_primes(100)
print("Twin prime pairs between 1 and 100:")
for pair in twin_prime_pairs:
    print(pair)
```

```python
# Loop: Generate and print the first 15 numbers of the Tribonacci sequence.
def tribonacci(n):
    tribonacci_numbers = []
    a, b, c = 0, 1, 1
    for _ in range(n):
        tribonacci_numbers.append(a)
        a, b, c = b, c, a + b + c
    return tribonacci_numbers
```

```python
# Function: Write a function that takes a string and compresses it using run-length encoding.
def run_length_encoding(s):
    if not s:
        return ""
    
    encoded_string = []
    count = 1
    
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            encoded_string.append(f"{s[i - 1]}{count}")
            count = 1
            
    encoded_string.append(f"{s[-1]}{count}")  # Append the last character and its count
    return ''.join(encoded_string)
```

```python
# If-Else: Write a program to accept a time (in 24-hour format) and determine if it is AM, PM, or Midnight/Noon.
def time_of_day(hour, minute):
    if hour < 0 or hour > 23 or minute < 0 or minute > 59:
        return "Invalid time format."
    
    if hour == 0 and minute == 0:
        return "Midnight"
    elif hour == 12 and minute == 0:
        return "Noon"
    elif hour < 12:
        return f"{hour:02}:{minute:02} AM"
    else:
        return f"{hour - 12:02}:{minute:02} PM"
```

```python
# Loop: Find and display all 3-digit Armstrong numbers.
def is_armstrong(num):
    digits = [int(d) for d in str(num)]
    return sum(d ** len(digits) for d in digits) == num
def find_armstrong_numbers():
    armstrong_numbers = []
    for num in range(100, 1000):
        if is_armstrong(num):
            armstrong_numbers.append(num)
    return armstrong_numbers
```

```python
# Function: Create a function to find the sum of the first &#39;n&#39; perfect numbers.
def is_perfect(num):
    return sum(i for i in range(1, num) if num % i == 0) == num
def sum_of_perfect_numbers(n):
    perfect_numbers = []
    count = 0
    num = 1
    while count < n:
        if is_perfect(num):
            perfect_numbers.append(num)
            count += 1
        num += 1
    return sum(perfect_numbers)
```

```python
# If-Else: Accept a date (dd-mm-yyyy) and check if it&#39;s valid considering leap years and month constraints.
def is_valid_date(day, month, year):
    if month < 1 or month > 12:
        return False
    if day < 1 or day > 31:
        return False
    
    # Check for months with less than 31 days
    if month in [4, 6, 9, 11] and day > 30:
        return False
    
    # Check for February
    if month == 2:
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            if day > 29:  # Leap year
                return False
        else:
            if day > 28:  # Non-leap year
                return False
    
    return True
```

```python
# Function: Write a function to convert a given number into its word equivalent (e.g., 123 to One Two Three).
def number_to_words(n):
    if n < 0:
        return "Negative numbers are not supported."
    
    words = {
        0: "Zero", 1: "One", 2: "Two", 3: "Three", 4: "Four",
        5: "Five", 6: "Six", 7: "Seven", 8: "Eight", 9: "Nine"
    }
    
    return ' '.join(words[int(digit)] for digit in str(n))
```
