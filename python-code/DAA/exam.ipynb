 1. Time Complexity of Basic Sorting Algorithms (Bubble, Selection, Insertion)
=================================================================================================
These algorithms demonstrate O(n^2) time complexity. The test function measures execution time for different list sizes to illustrate this.

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
```

```python
def analyze_basic_sorts():
    print("--- 1. Analysis of Basic Sorting Algorithms ---")
    sizes = [100, 500, 1000, 2000]
    algorithms = {
        "Bubble Sort": bubble_sort,
        "Selection Sort": selection_sort,
        "Insertion Sort": insertion_sort,
    }

    for name, sort_func in algorithms.items():
        print(f"\nAnalyzing {name}:")
        for size in sizes:
            # Generate a random list of numbers for each run
            test_list = [random.randint(1, size) for _ in range(size)]
            
            start_time = time.time()
            sort_func(test_list.copy()) # Use a copy to not affect other sorts
            end_time = time.time()
            
            print(f"  Input size {size}: {end_time - start_time:.6f} seconds")
    print("-" * 45 + "\n")
```

 2. Master Theorem for Analyzing Recursive Algorithms
=================================================================================================
The Master Theorem provides a "cookbook" method for solving recurrence relations of the form:
T(n) = a * T(n/b) + O(n^k) where a >= 1, b > 1.

```python
def master_theorem(a, b, k):
    """
    Analyzes a recurrence relation T(n) = a*T(n/b) + O(n^k) using the Master Theorem.
    Returns the time complexity in Big O notation.
    """
    print("--- 2. Master Theorem Analysis ---")
    print(f"Analyzing T(n) = {a}*T(n/{b}) + O(n^{k})")
    
    # Calculate log_b(a)
    log_b_a = -1
    if a > 0:
        import math
        log_b_a = math.log(a, b)

    if log_b_a == -1:
        return "Invalid input: 'a' must be >= 1."

    # Case 1: log_b(a) > k
    if log_b_a > k:
        print(f"Case 1 applies: log_{b}({a}) > {k}")
        return f"Result: O(n^log_{b}({a})) which is approx O(n^{log_b_a:.2f})"
        
    # Case 2: log_b(a) == k
    elif log_b_a == k:
        print(f"Case 2 applies: log_{b}({a}) == {k}")
        return f"Result: O(n^{k} * log n)"
        
    # Case 3: log_b(a) < k
    else:
        print(f"Case 3 applies: log_{b}({a}) < {k}")
        return f"Result: O(n^{k})"

def analyze_master_theorem_samples():
    """Provides sample problems for the Master Theorem."""
    # Sample 1: Merge Sort -> T(n) = 2*T(n/2) + O(n) -> a=2, b=2, k=1
    print(master_theorem(2, 2, 1))
    
    # Sample 2: Binary Search -> T(n) = 1*T(n/2) + O(1) -> a=1, b=2, k=0
    print(master_theorem(1, 2, 0))
    
    # Sample 3: Strassen's Matrix Multiplication -> T(n) = 7*T(n/2) + O(n^2) -> a=7, b=2, k=2
    print(master_theorem(7, 2, 2))
    print("-" * 32 + "\n")
```

 3. Square of a Large Integer using Divide and Conquer
=================================================================================================
This method avoids using built-in large number handling by implementing the Karatsuba
algorithm for multiplication, which is faster than the classical grade-school algorithm.
For a number X with n digits, split it into two halves: X = a * 10^(n/2) + b
X^2 = (a * 10^(n/2) + b)^2 = a^2 * 10^n + 2ab * 10^(n/2) + b^2

```python
def large_integer_square(num_str):
    """
    Calculates the square of a large integer (provided as a string)
    using a divide and conquer strategy.
    """
    n = len(num_str)
    
    # Base case for recursion
    if n == 0:
        return 0
    if n <= 4: # Use standard multiplication for small numbers
        return int(num_str) * int(num_str)

    # Divide the number into two halves
    half = n // 2
    a_str = num_str[:half]
    b_str = num_str[half:]
    
    # Recursively compute the squares of the halves and their product
    # Note: Using str() to handle potentially large intermediate results
    a = int(a_str)
    b = int(b_str)

    # Karatsuba's trick to reduce multiplications
    # z2 = a^2
    # z0 = b^2
    # z1 = (a+b)^2 - z2 - z0  (which is 2ab)
    z2 = large_integer_square(a_str)
    z0 = large_integer_square(b_str)
    z1 = large_integer_square(str(a + b)) - z2 - z0
    
    # Combine the results: z2 * 10^n + z1 * 10^(n/2) + z0
    power_of_10_n = (n - half) * 2
    power_of_10_half = n - half
    
    return z2 * (10**power_of_10_n) + z1 * (10**power_of_10_half) + z0

def test_large_square():
    print("--- 3. Square of a Large Integer ---")
    large_num = "12345678901234567890"
    print(f"Number: {large_num}")
    
    result = large_integer_square(large_num)
    print(f"Square (Divide and Conquer): {result}")
    
    # Verification using Python's built-in capabilities
    expected = int(large_num) ** 2
    print(f"Square (Built-in Python):   {expected}")
    print(f"Result is correct: {result == expected}")
    print("-" * 36 + "\n")
```

 4. Job Scheduling with Deadlines (Greedy Algorithm)
=================================================================================================
The goal is to maximize profit by scheduling jobs, each taking one unit of time.
The greedy strategy is to sort jobs by their profit in descending order. Then, for each job,
schedule it as late as possible before its deadline.

```python
def job_scheduling(jobs):
    """
    Schedules jobs to maximize profit using a greedy algorithm.
    Each job is a tuple: (job_id, deadline, profit)
    """
    # Sort jobs by profit in descending order
    jobs.sort(key=lambda x: x[2], reverse=True)
    
    max_deadline = max(job[1] for job in jobs)
    
    # Time slots for the schedule, initially empty
    schedule = [-1] * max_deadline
    total_profit = 0
    scheduled_jobs = []
    
    for job_id, deadline, profit in jobs:
        # Find a free slot for this job, starting from its deadline and moving backwards
        for i in range(min(max_deadline, deadline) - 1, -1, -1):
            if schedule[i] == -1: # If slot is free
                schedule[i] = job_id
                total_profit += profit
                scheduled_jobs.append(job_id)
                break
                
    return scheduled_jobs, total_profit

def test_job_scheduling():
    print("--- 4. Job Scheduling with Deadlines ---")
    # Format: (Job ID, Deadline, Profit)
    jobs = [
        ('J1', 2, 100), ('J2', 1, 19), ('J3', 2, 27),
        ('J4', 1, 25), ('J5', 3, 15)
    ]
    print(f"Jobs (ID, Deadline, Profit): {jobs}")
    
    scheduled_sequence, max_profit = job_scheduling(jobs)
    
    print(f"\nOptimal job sequence for maximum profit: {scheduled_sequence}")
    print(f"Maximum profit: {max_profit}")
    print("-" * 40 + "\n")
```

 5. Merge Sort and Time Complexity Analysis
=================================================================================================
Merge Sort is a classic divide-and-conquer algorithm with a time complexity of O(n log n).
It works by recursively splitting the array in half, sorting the halves, and then merging them.

```python
def merge_sort(arr):
    """Sorts an array using the Merge Sort algorithm."""
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        # Recursive calls to sort the two halves
        merge_sort(left_half)
        merge_sort(right_half)

        # Merging the sorted halves
        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i] < right_half[j]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
    return arr
```

```python
def analyze_merge_sort():
    """Analyzes the time complexity of Merge Sort."""
    print("--- 5. Merge Sort Analysis ---")
    print("Merge Sort has a time complexity of O(n log n) in all cases (worst, average, best).")
    print("This is because the array is always split in half, leading to log n levels of recursion,")
    print("and each level involves O(n) work to merge the subarrays.")
    
    sizes = [1000, 5000, 10000, 20000]
    print("\nMeasuring performance:")
    for size in sizes:
        test_list = [random.randint(1, size) for _ in range(size)]
        start_time = time.time()
        merge_sort(test_list.copy())
        end_time = time.time()
        print(f"  Input size {size}: {end_time - start_time:.6f} seconds")
    print("-" * 30 + "\n")
```

# 6. Quick Sort with Different Pivot Strategies
=================================================================================================
Quick Sort is another efficient divide-and-conquer sorting algorithm. Its performance
heavily depends on the choice of the pivot element. A good pivot splits the array into
roughly equal halves, while a bad pivot (e.g., the smallest or largest element) can
degrade performance to O(n^2).

```python
def quick_sort(arr, low, high, pivot_strategy='last'):
    """Sorts an array using Quick Sort with a specified pivot strategy."""
    if low < high:
        pi = partition(arr, low, high, pivot_strategy)
        quick_sort(arr, low, pi - 1, pivot_strategy)
        quick_sort(arr, pi + 1, high, pivot_strategy)

def partition(arr, low, high, pivot_strategy):
    """Partitions the array and returns the partition index."""
    pivot_index = -1
    
    # --- Pivot Selection ---
    if pivot_strategy == 'last':
        pivot_index = high
    elif pivot_strategy == 'first':
        # Swap first element with last to use the same partition logic
        arr[low], arr[high] = arr[high], arr[low]
        pivot_index = high
    elif pivot_strategy == 'random':
        rand_pivot_index = random.randint(low, high)
        arr[rand_pivot_index], arr[high] = arr[high], arr[rand_pivot_index]
        pivot_index = high
        
    pivot_value = arr[pivot_index]
    i = low - 1
    
    for j in range(low, high):
        if arr[j] <= pivot_value:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def compare_quick_sort_pivots():
    """Compares the performance of different Quick Sort pivot strategies."""
    print("--- 6. Quick Sort Pivot Strategy Comparison ---")
    size = 1000
    strategies = ['first', 'last', 'random']
    
    # Case 1: Random Array (Average Case)
    print("\nPerformance on a Random Array:")
    random_array = [random.randint(1, size) for _ in range(size)]
    for strategy in strategies:
        arr_copy = random_array.copy()
        start_time = time.time()
        quick_sort(arr_copy, 0, len(arr_copy) - 1, pivot_strategy=strategy)
        end_time = time.time()
        print(f"  '{strategy}' pivot: {end_time - start_time:.6f} seconds")

    # Case 2: Sorted Array (Worst-Case for 'first' and 'last' pivots)
    print("\nPerformance on a Sorted Array (Worst Case for first/last):")
    sorted_array = list(range(size))
    for strategy in strategies:
        arr_copy = sorted_array.copy()
        start_time = time.time()
        quick_sort(arr_copy, 0, len(arr_copy) - 1, pivot_strategy=strategy)
        end_time = time.time()
        print(f"  '{strategy}' pivot: {end_time - start_time:.6f} seconds")
    print("-" * 49 + "\n")
```

 7. Binary Search and Efficiency Analysis
=================================================================================================
Binary Search is a highly efficient searching algorithm with a time complexity of O(log n).
CRITICAL: It only works on SORTED arrays. If the array is unsorted, the algorithm
will not work correctly and may return incorrect results.

```python
def binary_search(arr, target):
    """
    Performs binary search on a sorted array to find a target value.
    Returns the index of the target or -1 if not found.
    """
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

def analyze_binary_search():
    print("--- 7. Binary Search Analysis ---")
    
    # Case 1: Sorted Array (Correct Use Case)
    sorted_array = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
    target = 23
    print(f"Sorted Array: {sorted_array}")
    print(f"Target: {target}")
    index = binary_search(sorted_array, target)
    print(f"Result (Sorted Array): Target found at index {index}")
    print("Efficiency on Sorted Array: O(log n). With each step, the search space is halved.")

    # Case 2: Unsorted Array (Incorrect Use Case)
    unsorted_array = [56, 12, 91, 5, 38, 2, 16, 72, 23, 8]
    print(f"\nUnsorted Array: {unsorted_array}")
    print(f"Target: {target}")
    index = binary_search(unsorted_array, target)
    print(f"Result (Unsorted Array): Index found is {index} (likely incorrect).")
    print("Efficiency on Unsorted Array: The algorithm is not applicable.")
    print("Binary search relies on the ordered property of the array to work. On an unsorted")
    print("array, its assumptions are violated, leading to incorrect results. You must sort")
    print("the array first (e.g., in O(n log n)) before using binary search.")
    print("-" * 35 + "\n")
```
