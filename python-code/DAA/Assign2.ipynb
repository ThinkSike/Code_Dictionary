## Implement the Master Theorem to analyze recursive algorithms and verify results with sample problems.

Accept values of 'a', 'b' and f(n) from the user. Calculate the values of k.
T(n) = aT(n/b) + f(n)
Apply 3 cases as per input values and calculate time complexity.

```python
import sympy as sp

def master_theorem(a, b, f_n):
    n = sp.Symbol('n')
    f_n = sp.sympify(f_n)
    log_ba = sp.log(a, b)

    if f_n.is_polynomial():
        degree = sp.degree(f_n)
    else:
        try:
            degree = sp.limit(sp.log(f_n)/sp.log(n), n, sp.oo)
        except:
            return "Cannot determine complexity - invalid f(n)"
    
    if degree < log_ba:
        return 
```
