```python
def karatsuba(x, y):
    x, y = str(x), str(y)
    if len(x) == 1 or len(y) == 1:
        return int(x) * int(y)
    n = max(len(x), len(y))
    n2 = n // 2
    a, b = int(x[:-n2] or 0), int(x[-n2:])
    c, d = int(y[:-n2] or 0), int(y[-n2:])
    ac = karatsuba(a, c)
    bd = karatsuba(b, d)
    ad_bc = karatsuba(a + b, c + d) - ac - bd
    return ac * 10**(2 * n2) + ad_bc * 10**n2 + bd

num = input("Enter 20-digit number: ")
print("Square:", karatsuba(num, num))
#input: 12345678901234567890
```
